```python
compare=dataset.loc[:,['Sales','Quantity','Discount','Profit']]
sns.pairplot(compare)
```

```python
correlation=compare.corr()
sns.heatmap(correlation,annot=True)
```

```python
correlation=data.corr()
sns.heatmap(correlation,annot=True)
```

```python
sns.relplot(x='Sales',y='Profit',hue='Category',data=data)
```
